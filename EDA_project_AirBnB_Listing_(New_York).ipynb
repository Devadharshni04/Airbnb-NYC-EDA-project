# **4.Data cleaning**

# Dealing with null values

```python
data.isnull().sum()
#dropping missing values rows
data.dropna(inplace=True)

#rechecking the shape of the dataset after dropping missing values
data.isnull().sum()
data.shape
```

# Duplicate handling

```python
#Ensure if any duplicates exists
data.duplicated().sum()

#dealing with duplicates
data[data.duplicated()]

#deleting all duplicates
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

# Dealing with data inconsistency






```python
#Type casting
#changing data types
data.dtypes
data["id"]=data["id"].astype(object)
data["host_id"]=data["host_id"].astype(object)
data.dtypes
```

# **`5.Data Analysis`**

# Univariate analysis

```python
#price distribution
data['price']

#Identifying outliers in price
df=data[data["price"]<1500]

sns.boxplot(data=df,x='price')
```

```python
#price distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',bins=100 )
plt.title("Price distribution")
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

# Feature engineering

```python
# ['price per bed']

df['price per bed']= df['price']/df['beds']
df.head()
# average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
# heat map - correlation of one variable with others for numerical column

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
```

### 💰 Estimated Annual Revenue Analysis

**To gain deeper business insights, we estimated the potential annual revenue for each Airbnb listing by multiplying the nightly price by the number of days available in a year (`availability_365`)**

```python
df['estimated_revenue'] = df['price'] * df['availability_365']
#analyse
revenue_by_neighbourhood = df.groupby('neighbourhood')['estimated_revenue'].mean().sort_values(ascending=False)
print(revenue_by_neighbourhood)
```

```python
#Visualize
revenue_by_neighbourhood.head(10).plot(kind='barh')
plt.xlabel('Average Estimated Annual Revenue')
plt.title('Top 10 Neighbourhoods by Average Revenue')
plt.show()
```
